```python
import pandas as pd
import re

def tslug(s):
    return re.sub('\W','',s).lower()
```

```python
seen_df = pd.read_csv("data/0_labelled_documents.csv")

oa_docs = pd.read_feather("../DFID-climate-health/lancet_update/2024/data/oa_docs.feather", columns=['id','title'])
oa_docs["tslug"] = oa_docs["title"].astype(str).apply(tslug)
```

```python

seen_df["tslug"] = seen_df["title"].astype(str).apply(tslug)
seen_oa = seen_df.rename(columns={"id":"nacsos_id"}).merge(oa_docs[["id","tslug"]], how="left")
seen_oa["oa_match"] = pd.notna(seen_oa["id"])
seen_oa["oa_match"].sum() / seen_oa.shape[0]
seen_oa = seen_oa.rename(columns={"content":"abstract","PY":"publication_year"})
seen_oa.to_csv("data/labels.csv", index=False)

seen_oa
```
